## Languages
In this notebook we will explore data about different languages
We are going to answer these questions
- Top spoken languages
- Origin overview

This is the dataset we are using https://www.kaggle.com/datasets/rajkumarpandey02/list-of-languages-by-total-number-of-speakers

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("languages_dataset.csv")
df.head()
# df.describe(include=[object])
```

```python
# cleanup data
def get_lang(long_form):
    if "(" in long_form:
        return long_form[:long_form.index("(")]
    return long_form
df['lang'] = df['Language'].apply(get_lang)


# total in numbers 
def calc_total(cur):
    x = cur.split(' ')
    if x[1] == 'billion':
        return float(x[0]) * 1000
    return float(x[0])

df['total_speakers_in_million'] = df['Total speakers(L1+L2)'].apply(calc_total)
df.head()
```
